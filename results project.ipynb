```python
import ast
import pandas as pd

# Split the input into lines
lines = results_text.strip().split('\n')

# Parse lines into dictionaries
results = []
for line in lines:
    # Extract the dictionary part and the score part from each line
    dict_part = line.split('with score')[0].strip('Completed: ').strip()
    score_part = line.split('with score')[1].strip()

    # Convert the dictionary part to a dictionary and the score part to a float
    param_dict = ast.literal_eval(dict_part)
    score = float(score_part)

    # Add the score to the dictionary
    param_dict['score'] = score
    results.append(param_dict)

# Convert results to DataFrame
df = pd.DataFrame(results)
df = df.rename({'negative': 'Negative Sampling'}, axis='columns')
print(df.head())  # Display the first few entries to verify
df['score'].max()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting function
def plot_parameter_impact(df, parameter):
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df, x=parameter, y='score', marker='o')
    plt.title(f'Impact of {parameter} on Score')
    plt.xlabel(parameter.capitalize())
    plt.ylabel('Score')
    plt.grid(True)
    plt.show()

# Plot impact of each parameter
plot_parameter_impact(df, 'vector_size')
plot_parameter_impact(df, 'window')
plot_parameter_impact(df, 'min_count')
plot_parameter_impact(df, 'Negative Sampling')
```
